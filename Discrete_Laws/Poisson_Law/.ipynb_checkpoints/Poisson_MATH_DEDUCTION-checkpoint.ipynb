# DISCRETE LAW (Poisson)

## Poisson

The Poisson is a discrete Law that gets a parameter $\lambda$ between $]0,+\infty[$


We will call from this moment the distribuition that have a Poisson distribution as the following


$ X_i \leadsto Poisson(\lambda) $


$X_i \leadsto P(\lambda)$



With the following formula: $P_{\lambda}(X=k)= \frac{\lambda^{k}e^{-\lambda}}{k!}$


where $k$ can assume the following values $N$

The ideia behind this random variable is that it expresses the probability of a given number of events occurring in a fixed interval of time if these events occur with a known constant mean rate and independently of the time since the last event.


Let's deduce some fundamental conceps about this Law:

## Expected value and Variance

Be $X_i \leadsto Poisson(\lambda) \$

By the definition:


<div style="margin-bottom: 30px;">
    
<br>
    
<center>
$E[X_i] = \sum_{k=1}^{\infty} x*P(x=k)$
</center
    
<br>
<br>
    
$\sum_{k=1}^{\infty} xP(x=k)  \iff \sum_{i=1}^{\infty} k\frac{\lambda^{k}e^{-\lambda}}{k!} \iff e^{-\lambda}\sum_{i=1}^{\infty} \frac{\lambda^{k}}{(k-1)!} \tag{1}$

<br>
    
<br>
    
<center> The fonction $f(x) = e^{x} = \sum_{k=0}^{\infty} \frac{x^{k}}{k!}$ is convergent $\forall x \in R $ thus all the operations that we gonna do after converge $\implies$ </center>
    
<br>
    
$ (e^{x})\prime = \sum_{k=0}^{\infty} k\frac{x^{k-1}}{k!} \iff (e^{x})\prime = e^{x} = \sum_{k=1}^{\infty} \frac{x^{k-1}}{(k-1)!} \implies xe^{x}= \sum_{k=1}^{\infty} \frac{x^{k}}{(k-1)!} \tag{2}$
    
<br>
    
Therefore we can apply this in the Eq(1) $\implies$
    
<br>
    
<center> $E[X_i] = e^{-\lambda}\sum_{i=1}^{\infty} \frac{\lambda^{k}}{(k-1)!} e^{-\lambda}\lambda e^{\lambda} \iff \lambda $ </center>
    
<br>

<center> $E[X_i] = \lambda \tag{X}$</center>
    
<br>
   
For $E[(X_i)^{2}]:$
    
<br>
    
<center>$E[(X_i)^{2}] = \sum_{k=1}^{\infty} x^{2}*P(x=k) \iff E[(X_i)^{2}] = \sum_{k=1}^{\infty} k^{2}\frac{\lambda^{k}e^{-\lambda}}{k!} \iff E[(X_i)^{2}] = e^{-\lambda}\sum_{k=1}^{\infty} k^{2}\frac{\lambda^{k}}{k!} \iff e^{-\lambda}\sum_{k=1}^{\infty} k\frac{\lambda^{k}}{(k-1)!}$</center>
    
<br>
<br>
We can use the expression (1) again to produce the following:
    
<br>
<br>
<center>$x^{2}e^{x}= \sum_{k=1}^{\infty} \frac{x^{k+1}}{(k-1)!} \implies $implies is to the Lebesgue sense , to the Riemann sense is one if and only if (unnecessary remark) 
<br><br>$ (x^{2}e^{x})\prime = \sum_{k=1}^{\infty} k\frac{x^{k}}{(k-1)!} + \sum_{k=1}^{\infty} \frac{x^{k}}{(k-1)!} \iff \sum_{k=1}^{\infty} k\frac{x^{k}}{(k-1)!} = e^{x}(x^{2}+2x) - xe^{x} = e^{x}(x^{2}+x)$ </center> 
<br>

<center>$E[(X_i)^{2}] = e^{-\lambda}\sum_{k=1}^{\infty} k\frac{\lambda^{k}}{(k-1)!} =  e^{-\lambda}e^{\lambda}(\lambda^{2}+\lambda)$</center>
    
<br>
<center>$E[(X_i)^{2}] = \lambda^{2}+\lambda$</center> 
<br>

Then
    
<br>
    
<center>$ Var[X_i] = E[(X_i)^{2}] - (E[X_i])^{2} = \lambda^{2}+\lambda - \lambda^{2}$</center>

<br>
    
<center>$ Var[X_i] = \lambda$</center>

<br>

</div>


## Poisson estimators (MM / MLE)

$\mu_{1}(1) = \overline{X_n}$

The first estimators commes from the method of moments:

<div style="margin-bottom: 30px;">

<br>
<center>$E[\overline{X_n}] = E[\frac{\sum_{k=1}^{n} X_i}{n}] = \frac{\sum_{k=1}^{n} E[X_i]}{n} = \frac{nE[X_i]}{n} = \lambda$</center>
<br>

<center>$\tilde{\lambda} = \overline{X_n}$</center>
    
<br>
    
</div>


The second estimator commes from the maximum likelihood estimation (MLE):

<div style="margin-bottom: 30px;">
    
    
<br>
    
    
Even with the variables are not independent we can right them as a produtory , we gonna optimize the fonction Ln because is easier to work 
    
<br>
    
<center>$Ln(X,\lambda)= ln(\prod_{i=1}^{n} \frac{\lambda^{x_i}e^{-\lambda}}{(x_i)!}) = ln(e^{-n\lambda}\lambda^{\sum_{i=1}^{n} x_i}\frac{1}{\prod_{i=1}^{n} (x_i)!}) = -n\lambda + ln(\lambda)\sum_{i=1}^{n} x_i -ln(\prod_{i=1}^{n} (x_i)!)$</center>    
<br>
    
<center>$\frac{\partial Ln(X,\lambda)}{\partial \lambda} =  -n +\frac{\sum_{i=1}^{n}x_i}{\lambda} = 0\implies$  </center> 
    
    
<br>    
<center> $\hat{\lambda} = \overline{X_n}$ </center>
<br>    
   
<br>
Then the estimators are the same and the following analyses will be the same 
<br>
    
</div>

### Bias of an estimator

<div style="margin-bottom= 30px;">
    <br>
    <center>$b_{\lambda}[\hat{\lambda}] = E[\hat{\lambda}] - \lambda = 0$</center>
    <br>
    <center>Therefore he is not bias $b_{\lambda}[\hat{\lambda}]=0$</center>
    <br>
</div>

### Risque of an estimator

<div style="margin-bottom= 30px;">
    <br>
    <center>$R_{\lambda}[\hat{\lambda}] = (b_{\lambda}[\hat{\lambda}])^{2} + Var[\hat{\lambda}]$</center>
    <br>
    <center>$Var[\hat{\lambda}]= Var[\frac{\sum_{i=1}^{n}X_i}{n}]= \frac{1}{n^{2}}\sum_{k=1}^{n}Var[X_i]= \frac{Var[X_i]}{n} = \frac{\lambda}{n}$</center>
    <br>
    <center>$R_{\lambda}[\hat{\lambda}] =  \frac{\lambda}{n}$</center>
    <br>
</div>

### Consistency of an estimator

Hypothèses et conditions

1. $(X_n)$ a collection of iid samples from a random variable 
2. $E[X_i]<\infty$

<div style="margin-botto: 30px">
    <br>
    <center>Using The strong law of large numbers(LLN) we get that :</center>
    <br>
    <center>${\displaystyle {\overline {X}}_{n}\ {\overset {\text{a.s.}}{\longrightarrow }}\ \lambda \qquad {\textrm {when}}\ n\to \infty .} $</center>
    <br>
    <center>And besides, we know that the function $f(x) = x$ is a continuous function then using the continuity we have the following : </center>
    <br>
    <center>${\displaystyle f({\overline {X}}_{n}\ ){\overset {\text{a.s.}}{\longrightarrow }}\ f(\lambda) \qquad {\textrm {when}}\ n\to \infty .} \iff {\displaystyle \hat{\lambda}{\overset {\text{a.s.}}{\longrightarrow }}{\lambda} \qquad {\textrm {when}}\ n\to \infty .} \implies {\displaystyle \hat{\lambda}{\overset {\text{Prob}}{\longrightarrow }}{\lambda} \qquad {\textrm {when}}\ n\to \infty .}$</center>
    <br>
    <center>Then $\hat{\lambda}$   is a consistent estimator. </center>
    <br>
</div>



### Convergence speed

Hypothèses et conditions

1. $(X_n)$ a collection of iid samples from a random variable 
2. $E[X_i]=\lambda$
3. $Var[X_i] = \lambda<\infty $


<div style="margin-botto: 30px">
    <br>
    <center>Using The central limit theorem (CLT) we get that :</center>
    <br>
    <center>${\displaystyle \frac{{\sqrt {n}}\left({\bar {X}}_{n}-\mu \right)}{\sqrt{Var[X_i]}}\mathrel {\overset {d}{\longrightarrow }} {\mathcal {N}}\left(0,1\right).}$</center>
    <br>
    <center>And we as know $\hat{\lambda} = \overline{X_n}$ and $E[X_i] = \lambda$ so , just substituing the values we have the following: : </center>
    <br>
    <center>${\displaystyle \frac{{\sqrt {n}}\left(\hat{\lambda}- \lambda \right)}{\sqrt{\lambda}}\mathrel {\overset {d}{\longrightarrow }} {\mathcal {N}}\left(0,1\right).}$</center>
    <br>
    <center>Then the convergence speed is the following</center>
    <br>
    <center>${\displaystyle {\sqrt {n}}\left(\hat{\lambda}- \lambda \right)\mathrel {\overset {d}{\longrightarrow }} {\mathcal {N}}\left(0,\lambda\right).}$</center>
    <br>
</div>
